```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('winequality-white.csv', delimiter = ";")
```

```python
plt.title('Histogram of Quality')
plt.xlabel('Quality')
plt.ylabel('Frequency')
plt.hist(df['quality'], edgecolor = 'black')
plt.tight_layout()
plt.show()
```
